```python
import pandas as pd

genomeTags = pd.read_csv('/content/drive/MyDrive/Movie_recommender/Data/genome-tags.csv')  

genomeScores = pd.read_csv('/content/drive/MyDrive/Movie_recommender/Data/genome-scores.csv') 

links = pd.read_csv('/content/drive/MyDrive/Movie_recommender/Data/links.csv') 

movies = pd.read_csv('/content/drive/MyDrive/Movie_recommender/Data/movies.csv') 

ratings = pd.read_csv('/content/drive/MyDrive/Movie_recommender/Data/ratings.csv') 

tags = pd.read_csv('/content/drive/MyDrive/Movie_recommender/Data/tags.csv') 
```

```python
genomeTags
```

```python
genomeScores
```

## Merge these two into one genomes dataset

```python
genomes = pd.merge(
    genomeScores,
    genomeTags,
    how="inner",
    on="tagId",
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
genomes
```

```python
links
```

## Merge links with genomes by movieId

```python
MovieMerge = pd.merge(
    links,
    genomes,
    how="inner",
    on="movieId",
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
MovieMerge
```

```python
movies
```

## Merge MovieMerge with movies by movieId

```python
MovieMerge = pd.merge(
    MovieMerge,
    movies,
    how="inner",
    on="movieId",
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
MovieMerge
```

```python
MovieMerge.to_csv('/content/drive/MyDrive/Movie_recommender/Data/MovieMerge.csv', index = False)
```
