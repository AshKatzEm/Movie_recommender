## Currently, the "genres" column of the "movies" dataframe combines all of a movie's genres as a single string. Let's see if we can split it up

```python
movies = movies.join(movies.genres.str.split("|", expand = True))
movies
```

ok, those column names are not working

```python
movies = movies.rename(columns = {0:'genre1', 1:'genre2',2:'genre3',3:'genre4',4:'genre5',5:'genre6', 6: 'genre7',7:'genre8',8:'genre9',9:'genre10'})
movies
```

## This method is actually pretty messy. Let's try one-hot encoding each genre. First, let's see how many genres we have

```python
movies['genre1'].unique()
```

```python
movies['genre2'].unique()
```

## It looks like every possible genre appears in the first genre column. Let's use that value list to create a column for each genre. 

```python
for i in movies['genre1'].unique():
  movies[i] = 0

movies
```

```python
genres = movies['genre1'].unique()
for i in genres:
  movies[i] = 0

movies

```

```python
for i in movies.index: # rows of the df
  for j in genres: # the array of possible values, not the df column
    if j in movies.genres[i]: # if a genre is contained in a movies genres
      movies[j][i] = 1 # store a 1 in that genres individual column
movies
```

## That looks like it worked, so let's clean up the previous attempt


```python
movies = movies.drop(labels=['genres', 'genre1', 'genre2', 'genre3','genre4','genre5','genre6', 'genre7','genre8','genre9','genre10'], axis=1)
```

# Ok, so let's move on. 
## In our final dataset, I think we only need the "Id"s. When a user enters a a tag, or a movie name, we can do a quick conversion with the genome or movies dataframe. In fact, let's create a "conversion" dataframe.

```python
conversion = pd.DataFrame(data = movies.title)
conversion['movieId'] = movies.movieId
conversion['tag'] = genomeTags.tag
conversion['tagId'] = genomeTags.tagId
conversion
```

## That looks like it will work, albeit a little akward. Now, let's start making our final dataset and then we'll remove the non-id columns

```python
RatingsMerge = pd.merge(
    ratings,
    tags,
    how="outer",
    on=["userId", 'movieId'],
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=False,
    suffixes=("_x", "_y"),
    copy=False,
    indicator=False,
    validate=None,
)
RatingsMerge
```

```python
print(len(tags))
print(len(RatingsMerge))
print(RatingsMerge.tag.isna().sum(), "+", len(tags), '=', RatingsMerge.tag.isna().sum() + len(tags) )
```

## so that merge seems to have worked, numerically at least

```python
print(len(ratings))
print(len(RatingsMerge))
print(RatingsMerge.rating.isna().sum(), "+", len(ratings), '=', RatingsMerge.rating.isna().sum() + len(ratings) )
print(ratings.rating.isna().sum())
print(RatingsMerge.userId.unique())
```

```python
tagmerge = pd.merge(
    tags,
    ratings,
    how="left",
    on=["userId", 'movieId'],
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=False,
    suffixes=("_x", "_y"),
    copy=False,
    indicator=False,
    validate=None,
)
tagmerge
```

```python
print(tagmerge.rating.isna().sum())
print(len(tagmerge.userId.unique()))
```

## That seems to be ok, let's try to roll with this

```python
tagmerge = pd.merge(
    tagmerge,
    movies,
    how="inner",
    on=['movieId'],
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=False,
    suffixes=("_x", "_y"),
    copy=False,
    indicator=False,
    validate=None,
)
tagmerge
```
