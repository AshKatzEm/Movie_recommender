This is actually pretty messy. Let's try one-hot encoding each genre. First, lets see how many genres we have

```python
movies['genre1'].unique()
```

```python
movies['genre2'].unique()
```

```python
for i in movies['genre1'].unique():
  movies[i] = 0

movies
```

```python
movies = pd.read_csv('/content/drive/MyDrive/Movie_recommender/Data/movies.csv') 
```

```python
genres = movies['genre1'].unique()
for i in genres:
  movies[i] = 0

movies

```

```python
for i in movies.index:#rows
  for j in genres:
    if j in movies.genres[i]:
      movies[j][i] = 1
movies
```

lets clean up now


```python
movies = movies.drop(labels=['genre1', 'genre2', 'genre3','genre4','genre5','genre6', 'genre7','genre8','genre9','genre10'], axis=1)
```
