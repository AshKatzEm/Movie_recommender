```python
import pandas as pd

MovieMerge = pd.read_csv('/content/drive/MyDrive/Movie_recommender/Data/MovieMerge.csv')  


ratings = pd.read_csv('/content/drive/MyDrive/Movie_recommender/Data/ratings.csv') 

tags = pd.read_csv('/content/drive/MyDrive/Movie_recommender/Data/tags.csv') 
```

```python
#manual merge
# for i in ratings.index:
#   if ratings.movieId[i] == MovieMerge.movieId:
RatingsMerge = pd.merge(
    ratings,
    tags,
    how="inner",
    on=["userId", 'movieId'],
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=False,
    suffixes=("_x", "_y"),
    copy=False,
    indicator=False,
    validate=None,
)
RatingsMerge
```

```python
RatingsMerge.to_csv('/content/drive/MyDrive/Movie_recommender/Data/RatingsMerge.csv', index = False)
```
